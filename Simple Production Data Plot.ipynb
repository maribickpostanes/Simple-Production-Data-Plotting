# Python Code

Visualizing production data in a simple plot using Python is essential in the oil and gas industry to identify production trends and patterns in real-time. The following code imports two Python libraries, namely matplotlib.pyplot and pandas, that offer function for data visualization, analysis, and manipulation. With these libraries imported, we can use their functionalities in our Python script to obtain the desired plot.

```python
# Import python libraries
import matplotlib.pyplot as plt
import pandas as pd
```

Matplotlib is a versatile python library for creating customizable data visualizations while Pandas is a python library that provides data manipulation and analysis functions.  Together, they are a powerful combination for efficient data analysis and visualization. After importing the required libraries, the code reads data from an Excel file with a .xlsx extension. This file format is commonly used to store and exchange data in spreadsheet form. The data used in this code to generate the desired plot is a production data from the Volve Field situated in the Norwegian North Sea.

```python
# Read Excel File
df = pd.read_excel('volve.xlsx')
```

As there are two categories of production data available (daily and monthly), I have selected the daily production. Here, I created specific lists in the code below that filters the original dataframe (df) object above and creates two new dataframe objects, x and y. I simply choose only two wells (15/9-F-12 H and 15/9-F-14 H) in the production data to be plotted in a graph at the same time. The filtering is done based on the 'Wellbore name' column such that only the rows matching '15/9-F-12 H' and '15/9-F-14 H' are kept in x and y, respectively. Thus, these new dataframe objects are now ready to be utilized in analyzing and visualizing the filtered data.

```python
# Create lists and filter the data frame to retrieve only two wells
x = df[df["Wellbore name"]=='15/9-F-12 H']
y = df[df["Wellbore name"]=='15/9-F-14 H']
```

Of course, the desired plot (please see the figure below) displaying the daily oil production data for two wells, namely '15/9-F-12 H' and '15/9-F-14 H' are now generated. It creates a figure and axes object that sets the size and layout of the plot using fig, ax = plt.subplots(figsize=[9,6]) command.

The x.plot and y.plot commands plot the data for each well with 'Date of Production' column as the x-axis and 'BORE_OIL_VOL' column as the y-axis. The line color and labels are defined using the c and label parameters, respectively.

The plot is made more informative by adding gridlines and a legend using the plt.grid(True) and plt.legend(title='Wellbore Name') commands. The title of the plot is set to 'DAILY OIL PRODUCTION' using plt.title('DAILY OIL PRODUCTION',fontsize=15).

```python
# Plot the data and do some customization
fig, ax = plt.subplots(figsize=[9,6])
x.plot(x='Date of Production',y='BORE_OIL_VOL',c='#425df5',label='NO 15/9-F-12 H',ax=ax)
y.plot(x='Date of Production',y='BORE_OIL_VOL',ylabel='Borehole Oil Volume\n(Sm3)',c='orange',label='NO 15/9-F-14 H',ax=ax)
plt.grid(True);plt.legend(title='Wellbore Name')
plt.title('DAILY OIL PRODUCTION',fontsize=15)
```

Finally, the plot that was desired is displayed.
